##  pairplot :-

* multiple plots at a time for the entair data

```python
sns.set_style('ticks')
sns.pairplot(iris,hue = 'species',diag_kind='kde',kind= 'scatter',palette='husl')
plt.show()
```

## strip plot

```python
sns.stripplot(x='species',y='sepal_length',data=iris)
```

## heatmap:-

```python
a = np.random.randn(10,20)
g = sns.heatmap(a)
plt.show()
```

```python
sns.heatmap()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(a,annot=True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(a,annot=False)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(a,annot=True,linewidth = 0.5)
plt.show()
```

```python
sns.heatmap(a,cmap='YLGnBu')
plt.show()
```
